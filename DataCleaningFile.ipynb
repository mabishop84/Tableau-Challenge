```python
frames=[df2102, df2103, df2104, df2105, df2106, df2107, df2108, df2109]
#frames=[df2101, df2102, df2103]
result = pd.concat(frames)
```

```python
result.info()
```

```python
result['ended_at']=pd.to_datetime(result['ended_at'])
result['started_at']=pd.to_datetime(result['started_at'])
```

```python
result['duration']=result['ended_at']-result['started_at']
```

```python
result['trip minutes']=result['duration'].dt.seconds.div(60).astype(int)
#result['trip minutes']=pd.to_numeric(result['trip minutes'])
```

```python
result['start hour']=result['started_at'].dt.hour.astype('category')
result['start day']=result['started_at'].dt.day.astype('category') 
result['Month']=result['started_at'].dt.month.astype('category')
result['weekday']=result['started_at'].dt.weekday.astype('category')
```
